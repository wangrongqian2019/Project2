# Denoise method 1 - 1

Here, we consider the minimization via total variation regularization
\begin{equation}
\min_{u} ||u-f||_{2}^{2}+\lambda||\nabla u||_{1}
\end{equation}
Solving it is not an apparent task, yet we can do the so called splitting method
\begin{equation}
\min_{u,p} ||u-f||_{2}^{2}+\lambda||p||_{1}+\mu||\nabla u -p||_{2}^{2}
\end{equation}
it follows that, for our final converging sequence, starting from $(u_{0},p_{0})$, we have two factors $(u_{k},p_{k})$, where for each iteration step, we need to compute two minimization problem.


[a] shrinkage thresholding step
\begin{equation}
p_{k+1}=argmin_{p}F_{k}^{1}(p)=argmin_{p}||\nabla {u_{k}}-p||_{2}^2+\lambda||p||_{1}
\end{equation}
[b] The $H^1$ regularization step
\begin{equation}
u_{k+1}=argmin_{u}F_{k}^{2}(u)=argmin_{u} ||u-f||_{2}^{2}+\mu||\nabla u-p_{k+1}||_{2}^2
\end{equation}

Therefore, we have two subproblems to consider. For the first one, simply perform the shrinkage on every term of $\nabla u_{k}$, namely.
\begin{equation}
p_{k+1}=T_{\lambda}(\nabla u_{k})
\end{equation}
Here $T_{\lambda}(x)$ is the shrinkage operator, which stands for

$$T_{\lambda}(x)=sgn(x)max\lbrace|x|-\lambda,0\rbrace$$

and $u_{k+1}$ is to be calculated via problem 3, which is about calculating the equation
\begin{equation}
2(u-f)-2\mu\Delta u+2\mu \nabla p_k=0
\end{equation}
gradient descent
$$   u_t =   (u-f) - \mu \Delta u +   \mu \nabla\cdot p_k $$
or can be written as
$$u_{k+1} = u_k +dt\cdot\left[ (u_{k}-f)-\mu (u_{xx,k}+u_{yy,k})+ \mu \nabla\cdot p_k\right] $$
where
\begin{equation}
\begin{split}
\nabla\cdot p_k =& \nabla\cdot T_\lambda(\nabla u) = \nabla\cdot\left(T_\lambda(u_{x+1} - u_{x}),T_\lambda(u_{y+1} - u_{y})\right)\\
=& T_\lambda(u_{x+1} - u_{x}) - T_\lambda(u_{x} - u_{x-1}) + T_\lambda(u_{y+1} - u_{y}) - T_\lambda(u_{y} - u_{y-1})
\end{split}
\end{equation}

```python
#defination of shrinkage function
def shrinkage (x, lambda0):
    lambdt = lambda0*np.ones(x.shape)
    return np.sign(x) * np.max(np.abs(x) - lambdt,0)
```

```python
# initialize parameter
lambd = 0.5
dt = 0.1
mu = 0.2
iteration = 100

I_t = nimg[:]
I_tplus = nimg[:]

# iteration
for t in range(iteration):
    for i in range(ny):
        for j in range(nx):
            iUp = i - 1
            iDown = i + 1
            jLeft = j - 1
            jRight = j + 1
            # boundary condition
            if (0 == i):
                iUp = i
            if (ny - 1 == i):
                iDown = i
            if (0 == j):
                jLeft = j 
            if (nx - 1 == j):
                jRight = j
            u_shrinkage = shrinkage(I_t[i][jRight] - I_t[i][j],lambd) - shrinkage(I_t[i][j] - I_t[i][jLeft],lambd) + shrinkage(I_t[iUp][j] - I_t[i][j],lambd) - shrinkage(I_t[i][j] - I_t[iDown][j],lambd)
            u_xx = I_t[i][jRight] + I_t[i][jLeft] - 2 * I_t[i][j]
            u_yy = I_t[iDown][j] + I_t[iUp][j] - 2 * I_t[i][j]
            I_tplus[i][j] = I_t[i][j] + dt * ((I_t[i][j] - nimg[i][j]) - mu * (u_xx + u_yy) + mu * u_shrinkage)
            
    I_t = I_tplus[:]
I_t11 = I_t
```

# Denoise method 1 - 2

\begin{equation}
\min_{u} \|u-f\|_{2}^{2}+\lambda\|\nabla u\|_{1}
\end{equation}
we can set 
$$ \|\nabla u\|_{1} = \int_{\Omega} \sqrt{u_{x}^{2}+u_{y}^{2}} d x d y$$
Euler-Lagrange equation: 
    $$2(u-f)-2\lambda\cdot\left[ \frac{\partial}{\partial x}\left(\frac{u_{x}}{\sqrt{u_{x}^{2}+u_{y}^{2}}}\right)+\frac{\partial}{\partial y}\left(\frac{u_{y}}{\sqrt{u_{x}^{2}+u_{y}^{2}}}\right)\right]=0$$
gradient descent: 
    $$u_{t}=\lambda\cdot \frac{u_{x}^{2} u_{y y}+u_{y}^{2} u_{x x}-2 u_{x} u_{y} u_{x y}}{\left(u_x^{2}+u_{y}^{2}\right)^{{3}/{2}}}-(u-f)$$
using forward difference method:
    $$u_{k+1}=u_{k}+dt\cdot \left[\lambda \cdot \frac{u_{x}^{2} u_{y y}+u_{y}^{2} u_{x x}-2 u_{x} u_{y} u_{x y}}{\left(u_x^{2}+u_{y}^{2}\right)^{{3}/{2}}}-(u_k-f) \right]$$

```python
# initialize parameter
lambd = 1.2      
dt = 0.2
ep = 1         #gradient regularization, otherwise denominator is 0
iteration = 100

I_t = nimg[:]
I_tplus = nimg[:]

# iteration
for t in range(iteration):
    for i in range(ny):
        for j in range(nx):
            iUp = i - 1
            iDown = i + 1
            jLeft = j - 1
            jRight = j + 1
            # boundary condition
            if (0 == i):
                iUp = i
            if (ny - 1 == i):
                iDown = i
            if (0 == j):
                jLeft = j 
            if (nx - 1 == j):
                jRight = j
            u_x = (I_t[i][jRight] - I_t[i][jLeft]) / 2.0
            u_y = (I_t[iDown][j] - I_t[iUp][j]) / 2.0
            u_xx = I_t[i][jRight] + I_t[i][jLeft] - 2 * I_t[i][j]
            u_yy = I_t[iDown][j] + I_t[iUp][j] - 2 * I_t[i][j]
            u_xy = (I_t[iDown][jRight] + I_t[iUp][jLeft] - I_t[iUp][jRight] - I_t[iDown][jLeft]) / 4.0
            num = u_yy * (u_x * u_x + ep) + u_xx * (u_y * u_y + ep) - 2 * u_x * u_y * u_xy
            den = pow(u_x * u_x + u_y * u_y + ep, 1.5)
            I_tplus[i][j] = I_t[i][j] + dt * (lambd * (num / den) - (I_t[i][j] - nimg[i][j]))
            
    I_t = I_tplus[:]
I_t12 = I_tplus[:]
```

# Denoise method 1 - 3

Here, we consider the minimization via total variation regularization
\begin{equation}
\min_{u} ||u-f||_{2}^{2}+||\nabla u||_{1}
\end{equation}
Solving it is not an apparent task, yet we can do the so called splitting method
\begin{equation}
\min_{u,p} ||u-f||_{2}^{2}+||p||_{1}+\frac{\alpha}{2}||\nabla u -p||_{2}^{2} +<\mu,p-\nabla u>
\end{equation}

[a] the u iteration
\begin{equation}
u_{k+1}=argmin_{u} ||u-f||_{2}^{2}+\frac{\alpha}{2}||\nabla u-p_{k+1}||_{2}^{2} -<\mu_{k},\nabla u>
\end{equation}
Then we have it's optimization condition
\begin{equation}
u_{k+1}=(I-\alpha\Delta u_{k+1})^{-1}(f-\alpha\nabla\cdot p_{k}-\nabla\mu_{k})
\end{equation}
[b]  the shrinkage for $p_{k}$
denote by $T_{\lambda}$ the shrinkage operator, we have
\begin{equation}
p_{k+1}=argmin_{p}\frac{\alpha}{2}||\nabla u_{k+1}-p||_{2}^2+<\mu_{k},p>+\lambda||p||_{1}
\end{equation}
Namely, by inserting the independent term and make it square, we have
\begin{equation}
p_{k+1}=T_{{1}/{\alpha}}(\nabla u_{k+1}+\frac{1}{\alpha}\mu_{k})
\end{equation}
here, the shrinkage reads
\begin{equation}
T_{\lambda}(x)=sgn(x)max\lbrace|x|-\lambda,0\rbrace
\end{equation}

[c] finally the $\mu $iteration
\begin{equation}
\mu_{k+1}=\mu_{k}+\alpha(p_{k+1}-\nabla u_{k+1})
\end{equation}


```python
N=200;
u1=zeros(N,N);
du1=zeros(N,N,2);
f=100.0*ones(N,N);
i=75;
j=75;
while i<151
    while j<151
     f(i,j)=150;
    j=j+1;
    end
    i=i+1;
    j=75;
end
nmean=0.0;
nsigma=12.0;
nimg=normrnd(nmean,nsigma,N)+f;
u=nimg;
epsilon=0;
p=zeros(N,N,2);
mu=zeros(N,N,2);
np=zeros(N,N);
nmu=zeros(N,N);
a=0.0001;
dt=1;
i=1;
j=1;
k=0;
t=0;
M=6;
while t<M
while i<N+1
    while j<N+1
        iup=i-1;
        idown=i+1;
        jleft=j-1;
        jright=j+1;
        if 1==i
            iup=i;
        end
        if N==i
            idown=i;
        end
        if 1==j
            jleft=j;
        end
        if N==j
            jright=j;
        end
        %% the u step
        u_yy=u(i,jright)+u(i,jleft)-2*u(i,j);
        u_xx=u(idown,j)+u(iup,j)-2*u(i,j);
        np(i,j)=0.5*(p(i,jright,2)-p(i,jleft,2)+p(iup,j,1)-p(idown,j,1));
        nmu(i,j)=0.5*(mu(i,jright,2)-mu(i,jleft,2)+mu(iup,j,1)-mu(idown,j,1));
        u1(i,j)=u(i,j)+dt*((u(i,j)-f(i,j))-a*(u_xx+u_yy)-nmu(i,j)+a*np(i,j));
        j=j+1;
    end
    i=i+1;
    j=1;
end
while i<N+1
    while j<N+1
        iup=i-1;
        idown=i+1;
        jleft=j-1;
        jright=j+1;
        if 1==i
            iup=i;
        end
        if N==i
            idown=i;
        end
        if 1==j
            jleft=j;
        end
        if N==j
            jright=j;
        end
        # the p step
        du1(i,j,1)=0.5*(u1(iup,j)-u1(idown,j));
        du1(i,j,2)=0.5*(u1(i,jright)-u1(i,jleft));
        p(i,j,1)=shrinkage(1/a,du1(i,j,1)+(1/a)*mu(i,j,1));
        p(i,j,2)=shrinkage(1/a,du1(i,j,2)+(1/a)*mu(i,j,2));
        # the mu step
        mu(i,j,1)=mu(i,j,1)+a*(p(i,j,1)-du1(i,j,1));
        mu(i,j,2)=mu(i,j,2)+a*(p(i,j,2)-du1(i,j,2));
        j=j+1;
    end
    i=i+1;
    j=1;
end
i=1;
j=1;
while i<N+1
    while j<N+1
        u(i,j)=u1(i,j);
        j=j+1;
    end
    i=i+1;
    j=1;
end
t=t+1;
i=1;
j=1;
end
imshow(u,[75,200]);
```

# Denoise method 2

$$\min _{u}\|u-f\|_{2}^{2}+\lambda\|u\|_{2}^{2} $$
solves the equation
$$2(u-f)+2 \lambda u=0 $$
hence
$$u=\frac{1}{1+\lambda} f$$

# Denoise method 3

$$\min _{u}\|u-f\|_{2}^{2}+\lambda\|\nabla u\|_{2}^{2}$$
Euler-Lagrange equation:
$$2(u-f)-2 \lambda \Delta u=0$$
gradient descent:
$$u_{t}=\lambda\left(u_{x x}+u_{y y}\right)-(u-f)$$
using forward difference method:
$$u_{k+1}=u_{k}+d t \cdot\left[\lambda\left(u_{x x, k}+u_{y y, k}\right)-\left(u_{k}-f\right)\right]$$

```python
# initialize parameter
lambd = 0.5
dt = 0.1
iteration = 100
    
I_t = nimg[:]
I_tplus = nimg[:]

# iteration
for t in range(iteration):
    for i in range(ny):
        for j in range(nx):
            iUp = i - 1
            iDown = i + 1
            jLeft = j - 1
            jRight = j + 1
            # boundary condition
            if (0 == i):
                iUp = i
            if (ny - 1 == i):
                iDown = i
            if (0 == j):
                jLeft = j 
            if (nx - 1 == j):
                jRight = j
            u_xx = I_t[i][jRight] + I_t[i][jLeft] - 2 * I_t[i][j]
            u_yy = I_t[iDown][j] + I_t[iUp][j] - 2 * I_t[i][j]
            I_tplus[i][j] = I_t[i][j] + dt * (lambd * (u_xx + u_yy) - (I_t[i][j] - nimg[i][j]))
            #I_tplus[i][j] = lambd * (u_xx + u_yy) + nimg[i][j]
            
    I_t = I_tplus[:]
I_t3 = I_t
```
